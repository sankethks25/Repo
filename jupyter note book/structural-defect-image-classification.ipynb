```python
class ImageProcessing:
    
    def clean_data(self,fname,target_size=(60, 60)):
        img = plt.imread(fname)
        img = cv2.resize(img, target_size) 
        # Normalization
        img = img/255.0
        return img 
    
    def set_X_Y(self,img_path = "E:/STRUCTURAL-DEFECT-IMAGES/ALL-DEFECT-IMAGES"):
        data_dir = Path(r"{img_path}")
        X = []
        y = []
        counter = 0
        for file in tqdm(file_names):
            if counter <= 10000:
                head, tail = os.path.splitext(file)
                if tail in [".jpg", ".jpe", "jpeg"]:
                    v = head.split("_")
                    service_num = int(v[0])
                    order_num = int(float(v[1]))
                    cat = df[(df["Service Order Number"]==snum) & (df["Item No. Order Doc."]==onum)]['Category'].values
                    if cat:   
                        img = self.clean_data(data_dir / file)
                        if img.shape == (60, 60, 3):
                            X.extend([img])
                            y.extend([cat[0]])
                        else: print(f"Shape miss matched file: {file}")

                    else:
                        print(file)
            else:
                break

            counter+=1
            return X,y
        
    def save_data(self,X,y,file_path):
        spath = Path(r""+file_path)
        joblib.dump(X,f"{spath}/X.pickle")
        joblib.dump(y,f"{spath}/y.pickle")
        print(f"Data has been saved to {spath}")
        
    def load_saved_data(self):
        file_path = Path(r"C:\Users\sanketh\Desktop\JUPYTER\DEEP-LEARNING\features_labels")
        X = joblib.load(f"{file_path}/X.pickle")
        y=joblib.load(f"{file_path}/Y.pickle")
        return X,y
        
```

```python
inputs = Input(shape=X[0].shape,name = "input_layer")

x = Conv2D(filters=256,kernel_size=(5,5),strides=(1,1),padding="valid",activation="relu")(inputs)
x = MaxPool2D(pool_size=(2,2),strides=(1,1))(x)
x = BatchNormalization()(x)

x = Conv2D(filters=128,kernel_size=(5,5),strides=(1,1),padding="valid",activation="relu")(x)
x = MaxPool2D(pool_size=(2,2),strides=(1,1))(x)
x = BatchNormalization()(x)

x = Dense(activation="relu",units=528,)(x)
x = Flatten()(x)

```

```python
outputs = Dense(units=29,activation="softmax",)(x)
```

```python
model = Model(inputs = inputs,outputs = outputs)
```

```python
def callbacks():
    
    earlystopping = EarlyStopping(monitor="val_loss", patience=5, mode="min", verbose=1, restore_best_weights=True)
    checkpoint = ModelCheckpoint("model.hdf5", monitor='val_loss', verbose=1, save_best_only=True,  mode='min', save_freq='epoch')
    lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=3, verbose=1, mode='min', min_delta=0.0001, cooldown=0, min_lr=0)
    return [earlystopping, checkpoint, lr]
```

```python
def encode_y():
    new_y = []
    dic = {} 
    filtered_y = np.unique(y)
    
    for i in range(len(filtered_y)):
        dic[filtered_y[i]] = i
    for category in y:
        new_y.append(dic[category])
    return new_y
new_y=encode_y()
```
