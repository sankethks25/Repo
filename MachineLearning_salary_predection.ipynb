```python
#here we are not using sklearn library but we are implementing linear regression algorithm
# M = + X + C
# M = (covariance of variables) / (variance of independent variable)
# C = (mean of dependent variable) - M * (mean of independent variable)
# VARIANCE = sum ( (atteribute_value - attribute_mean) **2)
#COVARIANCE = ((independent_var, independent_var_mean, dependent_var, dependent_var_mean)

import pandas as pd
import numpy as np
M = 0
N = 0

data = pd.read_csv("C:\pycharm\ML\pythonProject\salary_predection\Salary_Data.csv")
df = pd.DataFrame(data)
df.head()

#mean of experience and salary
exp_mean = sum( df['YearsExperience']) / float(len(df['YearsExperience']))
sal_mean = sum(df["Salary"]) / float(len(df["Salary"]))

#  variance function

def variance(value,mean):
    return sum([ (val - mean)**2 for val in value ])

#covariance function

def covariance(exp, exp_mean, salary, sal_mean):
    covariance =0.0
    for i in range(len(exp)):
        covariance = covariance + (exp[i] - exp_mean) * (salary[i] - sal_mean)
    return covariance
#calling variance and covariance
        
v_exp = variance(df['YearsExperience'], exp_mean)
v_sal = variance(df['Salary'], sal_mean)
c_exp_sal = covariance(df['YearsExperience'], exp_mean, df['Salary'], sal_mean)

M = (c_exp_sal / v_sal)
C = sal_mean - M *(exp_mean)

salary = M + 5 + C
salary








```

```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import linear_model
import seaborn as sb

data = pd.read_csv("C:\pycharm\ML\pythonProject\salary_predection\Salary_Data.csv")
df = pd.DataFrame(data)

%matplotlib inline
plt.xlabel('YearsExperience')
plt.ylabel('Salary')

plt.scatter(df.YearsExperience,df.Salary,color = 'red', marker = '+')

x =  np.array(df[["YearsExperience"]]) # np.array(df.drop(["Salary"],1))
y =  np.array(df[["Salary"]])                                  #df.iloc[:, 1].values
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 1/3, random_state = 0)
reg = linear_model.LinearRegression()
reg.fit(x_train,y_train)
y_pre = reg.predict(x_test)


p = reg.predict([[1000]])










```
